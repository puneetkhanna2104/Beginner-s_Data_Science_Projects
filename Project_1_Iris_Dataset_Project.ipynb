##Importing Packages

```
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

###Loading Data

```
iris = load_iris()
```

```
# Data -- Features 
x = iris.data
print(x)
```

```
# features  names 
feature_names = iris.feature_names
feature_names
```

```
# Data ---labels 
y = iris.target
```

```
# target(flower) names 
label_names = iris.target_names
print(label_names)
```

### Making DataFrame 

```

df = pd.DataFrame(iris.data, columns = feature_names)

# Adding labels(species) column into the dataframe 

df['species'] = iris.target
```

```
df.head()
```

```
df.shape
```

```
## DataFrame for Analyzing Data 

df1=df
df1['species'].replace([0,1,2],['setosa', 'versicolor', 'virginica'],inplace = True)
df1.head()
```

## Dividing the dataset into  training and testing data 

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df[iris.feature_names],df['species'],random_state = 0, shuffle =True)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## KNN Classifier 


```
from sklearn.neighbors import KNeighborsClassifier

# Loading the Classifier 
model = KNeighborsClassifier(n_neighbors = 3)

print(model)
```

###Training the Model 

```
model.fit(x_train, y_train)
```

###Testing the model on test data 

```
# Predicting using trained model 

predictions = model.predict(x_test)
```

```
# Calculating the accuracy of the model 
from sklearn.metrics import accuracy_score

acc= accuracy_score(y_test,predictions)

print('Accuracy on test data:', acc)
```

##Decision Tree Classifier 

```
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=0)
```

```
model.fit(x_train, y_train)
```

```
## Accuracy of the model 
model.score(x_test,y_test)
```

```
# Predicting using trained model 

predictions = model.predict(x_test)
```

```
# Calculating the accuracy of the model 
from sklearn.metrics import accuracy_score

acc= accuracy_score(y_test,predictions)

print('Accuracy on test data:', acc)
```
